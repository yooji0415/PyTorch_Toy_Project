```python
# 국민청원 사이트 크롤링 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time

# 크롤링 데이터를 저장하기 위한 데이터 프레임
result = pd.DataFrame()
# 2020년 국민청원 이용
for i in range(584274, 595226):
    URL = "http://www1.president.go.kr/petitions/" + str(i)
    
    # get 결과 저장
    response = requests.get(URL)
    # html 코드 분리
    html = response.text
    # 데이터 추출 Parsing 진행
    soup = BeautifulSoup(html, 'html.parser')
    # 제목과 참여인원 정보를 분리 
    title = soup.find('h3', class_= 'petitionsView_title')
    count = soup.find('span', class_= 'counter')
    
    # 청원 내용 분리
    for content in soup.select('div.petitionsView_write > div.View_write'):
        content
    
    # 청원 정보를 분리한다
    # 순서대로 카테고리, 청원시작, 청원마감, 청원인 순이다.
    a = []
    for tag in soup.select('ul.petitionsView_info_list > li'):
        a.append(tag.contents[1])
        
    if len(a) != 0:
        df1 = pd.DataFrame({
            'start': [a[1]],
            'end': [a[2]],
            'category': [a[0]],
            'count': [count.text],
            'title': [title.text],
            'content': [content.text.strip()[0:13000]]
        })
        # 청원을 데이터 프레임으로 생성 후 이를 result에 저장
        result = pd.concat([result, df1])
        # result 데이터 프레임의 인덱스 설정
        result.index = np.arange(len(result))
        
    if i % 60 == 0:
        print("Sleep 90 seconds. Count: " + str(i) +
              ", Local Time :" + time.strftime('%Y-%m-%d', time.localtime(time.time())) +
              " " + time.strftime("%X", time.localtime(time.time())) +
              ", Data Length: " + str(len(result)))
        time.sleep(90)
```

```python
# 데이터 확인
print(result.shape)
df = result
df.head()
```

```python
# 데이터 전처리
import re

# 띄어쓰기 관련 글자들을 모두 공백으로 전환
def remove_white_space(text):
    text = re.sub(r'[\t\r\n\f\v]', ' ', str(text))
    return text

# 한글과 숫자가 아닌 글자들을 모두 공백으로 전환
# 영어를 지운 이유는 한글 word embedding vector를 생성하기 위함
def remove_special_char(text):
    text = re.sub('[^ ㄱ-ㅣ 가-힣 0-9]', ' ', str(text))
    return text

df.title = df.title.apply(remove_white_space)
df.title = df.title.apply(remove_special_char)

df.content = df.content.apply(remove_white_space)
df.content = df.content.apply(remove_special_char)
```

```python
# 토크나이징 및 변수 생성
# 형태소 분석기 패키지 konlpy 사용
import os
!pip install --user JPype1-1.1.2-cp39-cp39-win_amd64.whl

from konlpy.tag import Okt

okt = Okt()

#청원 제목의 경우 형태소 단위로 토크나이징
df['title_token'] = df.title.apply(okt.morphs)
# 청원 내용의 경우 명사 단위로 토크나이징
df['content_token'] = df.title.apply(okt.nouns)
```

```python
# 파생변수 생성
df['token_final'] = df.title_token + df.content_token
# 참여 인원에 있는 , 마크를 제거하고 int 형으로 바꿔준다
df['count'] = df['count'].replace({',': ''}, regex = True).apply(lambda x: int(x))
# 바뀐 것을 확인
print(df.dtype)
```

```python
df_drop = df[['token_final', 'label']]
```

```python
# word embedding
# 단어를 숫자로 변환하여 딥러닝 모델 학습이 가능하게 만든다
# 중심 토큰을 통해 주변 토큰을 맞추는 Skip-Gram Word2Vec 을 사용할 예정

# 자연어 처리에 사용되는 모델중 Word2Vec을 가져온다
from gensim.models import Word2Vec

embedding_model = Word2Vec(df_drop['token_final'], # embedding vector를 만들 데이터 원본
                           sg = 1,                 # 모델 구조 옵션 선택으로 1은 Skip-Gram, 0은 CBOW 
                           size = 100,             # embedding vector의 사이즈를 정한다
                           window = 2,             # 문맥 파악에 필요한 앞 뒤 토큰 수를 정한다
                           min_count = 1,          # 일정 횟수 이상 등장하지 않는 토큰은 제외한다
                           workers = 4             # 병렬 프로세서의 수를 지정한다
                          )

# 임베딩 모델에 몇개의 토큰이 저장되어 있는지 확인
print(embedding_model)

# 음주운전과 유사한 단어와 백터 값을 출력해본다
model_result = embedding_model.wv.most_similar("음주운전")
print(model_result)
```

```python
# 임베딩 모델 저장
from gensim.models import KeyedVectors

embedding_model.wv.save.word2vec_format('data/petitions_tokens_w2v')

loaded_model = KeyedVectors.load_word2vec_format('data/petitions_tokens_w2v')

model_result = loaded_model.most_similar("음주운전")
```

```python
# 데이터 셋 분할 및 저장
from numpy.random import RandomState

rng = RandomState()

# 80% 는 train 20% 는 val 로 저장
# 데이터 수가 적기 때문이다
tr = df_drop.sample(frac= 0.8, random_state= rng)
val = df_drop.loc(~df_drop.index.isin(tr.index))

tr.to_csv('data/train.csv', index= False, encoding= 'utf-8-sig')
val.to_csv('data/validation.csv', index= False, encoding= 'utf-8-sig')
```
